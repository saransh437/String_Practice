```python
# 9. Check whether a vehicle number plate (input) contains both digits and alphabets.
a = input('Enter a vehicle number: ')

has_alpha = False
has_digit = False

for i in a:
    if i.isalpha():
        has_alpha = True
    elif i.isdigit():
        has_digit = True
    else:
      pass
if has_alpha and has_digit:
    print("Valid")
else:
    print("Invalid")
```

```python
# 9. Write a program to classify the string:
# ○ Only alphabets → "Alphabetic"
# ○ Only digits → "Numeric"
# ○ Alphanumeric (mix of both) → "Alphanumeric"
a = input('Enter a string:')

if a.isalpha():
  print('alphabetic')
elif a.isnumeric():
  print('Numeric')
elif a.isalnum():
  print('Alphanumeric')
else:
  pass
```
