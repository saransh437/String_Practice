```python
# 1. Vowel-based Word Count
# Count how many words in a sentence start AND end with a vowel
# (case-insensitive).
a = input('ENter a string:')
count = 0
vow = ('a','e','i','o','u')
word = a.split()
for i in word:
  if i.startswith(vow) and i.endswith(vow):
    count = count + 1
print(count)
```

```python
# 2. Palindrome Words
# Print all palindrome words found in a given sentence.
# ○ Example: "mom and dad went to civic center" -> output: mom dad civic
a = input('ENter a string:')
b = a.split()
for i in b:
   if i==i[::-1]:
      print(i)


```

```python
# 3. Title Case Without split() or title()
# Convert an entire sentence into Title Case using loops only.
sen = input('Enter a sentence')
b = sen.split()
c = ""
for ch in b:
    c = c +" "+ ch.capitalize()
print(c)
```

```python

# 4. Prime Digits Count
# Take a numeric string and count how many digits are prime
a = input("Enter a numeric string: ")

count = 0
prime_digits = ('2', '3', '5', '7')

for ch in a:
    if ch in prime_digits:
        count += 1

print("Prime no", count)

```

```python
# 5. Sum of Digits in Alphanumeric String
# Add only the digits from the input string.
# ○ Example: "a1b4c2" -> Output: 7
a = input('Enter a number:')
count =0
for i in a:
  if i.isdigit():
    count = int(i)+count
print(count)
```

```python
# 6. Reverse String Without Slicing
# Reverse a string using loops only.
sen = input("Enter a number:")
chr = ""
for i in sen:
  chr =  i +chr

print(chr)
```
